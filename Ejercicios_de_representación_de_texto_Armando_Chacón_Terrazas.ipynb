3. Función auxiliar para imprimir dimensiones y tokens

```python
def imprimir_info_matriz(nombre, vectorizer, X, n_tokens=10):
    """
    Imprime:
    - nombre de la matriz
    - dimensiones (n_docs, n_tokens)
    - primeros n_tokens del vocabulario y sus índices
    """
    print(f"\n=== {nombre} ===")
    print(f"Dimensiones de la matriz: {X.shape}")  # (n_documentos, tamaño_vocabulario)

    feature_names = vectorizer.get_feature_names_out()

    print(f"Primeros {n_tokens} tokens del vocabulario y sus índices:")
    for idx, token in enumerate(feature_names[:n_tokens]):
        print(f"Índice {idx}: '{token}'")
```

4. Función para comparar una fila Count vs TF‑IDF

```python
def comparar_fila(i, vectorizer_count, X_count, vectorizer_tfidf, X_tfidf, n_mostrar=15):
    """
    Muestra, para el documento i, los tokens que aparecen en ese documento
    y sus valores en CountVectorizer (conteos) y TfidfVectorizer (pesos TF‑IDF).
    """
    print(f"\n=== Comparación documento {i} (Count vs TF-IDF) ===")
    row_count = X_count[i].toarray().flatten()
    row_tfidf = X_tfidf[i].toarray().flatten()
    tokens = vectorizer_count.get_feature_names_out()  # mismo vocabulario

    print("Token\tCount\tTF-IDF")
    mostrados = 0
    for tok, c, t in zip(tokens, row_count, row_tfidf):
        if c > 0:  # solo tokens que aparecen en el doc
            print(f"{tok}\t{c}\t{t:.4f}")
            mostrados += 1
            if mostrados >= n_mostrar:
                break
```

# **EJERCICIO 1: TEXTO CRUDO (SIN PREPROCESAR MANUAL)**



```python
# CountVectorizer crudo
# Usamos lowercase=True y strip_accents='unicode' como se sugiere en la clase.
vectorizer_count_raw = CountVectorizer(
    lowercase=True,
    strip_accents='unicode'
)

X_count_raw = vectorizer_count_raw.fit_transform(corpus_es)

imprimir_info_matriz("CountVectorizer - texto crudo", vectorizer_count_raw, X_count_raw)

# TfidfVectorizer crudo

vectorizer_tfidf_raw = TfidfVectorizer(
    lowercase=True,
    strip_accents='unicode'
)

X_tfidf_raw = vectorizer_tfidf_raw.fit_transform(corpus_es)

imprimir_info_matriz("TfidfVectorizer - texto crudo", vectorizer_tfidf_raw, X_tfidf_raw)


# Comparación numérica (doc 0)
comparar_fila(
    i=0,
    vectorizer_count=vectorizer_count_raw,
    X_count=X_count_raw,
    vectorizer_tfidf=vectorizer_tfidf_raw,
    X_tfidf=X_tfidf_raw,
    n_mostrar=15
)
```

**CountVectorizer** genera una matriz de frecuencias absolutas (conteos de palabras por documento).

**TfidfVectorize**r genera una matriz con pesos TF-IDF, donde se resalta la relevancia de términos poco frecuentes en el corpus y se penalizan los muy comunes.

En el ejercicio 1 ambas matrices (**CountVectorizer y TfidfVectorizer**) tienen las mismas dimensiones (30, 127) y comparten exactamente el mismo vocabulario, por lo que los índices y los tokens aparecen iguales en ambos casos.
Esto ocurre porque los dos métodos parten del mismo modelo de bolsa de palabras: mismos 30 documentos y mismo conjunto de 127 términos extraídos del corpus.

La diferencia está en los valores numéricos de cada celda de la matriz. En la matriz de CountVectorizer cada posición contiene un entero que indica cuántas veces aparece ese token en el documento (frecuencia absoluta, por ejemplo 0, 1, 2, 3…). En cambio, en la matriz de TfidfVectorizer cada posición contiene un valor decimal que corresponde al peso TF‑IDF del token, que combina su frecuencia en el documento con su rareza en todo el corpus (por ejemplo 0.4136, 0.3779, 0.2920, etc.).

Así, aunque la forma de las matrices y el vocabulario sean idénticos, CountVectorizer solo cuenta apariciones, mientras que TfidfVectorizer reescala esas apariciones: asigna pesos bajos a palabras muy frecuentes en todo el corpus (como “la”, “los”, “me”) y pesos más altos a términos más raros y representativos (como “increíbles”, “guion”, “ambientación”).

# **EJERCICIO 2: TEXTO PREPROCESADO**

```python
# Definir preprocesamiento base.
# Lista sencilla de stopwords en español.
# Debemos proveer nuestra lista.
stopwords_es = {
    "el", "la", "las", "los", "un", "una", "unos", "unas",
    "de", "del", "al", "a", "y", "o", "u", "que", "como",
    "me", "mi", "mis", "tu", "tus", "su", "sus",
    "es", "fue", "son", "era", "ser", "soy", "eres",
    "en", "por", "para", "con", "sin", "lo",
    "muy", "pero", "no", "si", "ya", "hasta",
    "más", "mas", "menos", "nada", "todo", "toda", "todas", "todos",
    "esta", "este", "estos", "estas", "esa", "ese", "esos", "esas",
    "aqui", "alli", "ahi", "alla",
    "algunos", "algunas", "ninguno", "ninguna"
}

def quitar_acentos(texto):
    """
    Elimina acentos usando normalización unicode NFKD.
    """
    texto_norm = unicodedata.normalize("NFKD", texto)
    return "".join(c for c in texto_norm if not unicodedata.combining(c))

def preprocesar_texto(texto):
    """
    - Convierte a minúsculas.
    - Elimina acentos.
    - Elimina caracteres que no sean letras ni espacios.
    - Elimina stopwords en español.
    """
    # Minúsculas
    texto = texto.lower()

    # Quitar acentos
    texto = quitar_acentos(texto)

    # Quitar signos de puntuación y otros símbolos (dejamos solo letras y espacios)
    texto = re.sub(r"[^a-zñ\s]", " ", texto)

    # Tokenizar por espacios
    tokens = texto.split()

    # Eliminar stopwords
    tokens_limpios = [t for t in tokens if t not in stopwords_es]

    # Volver a unir
    return " ".join(tokens_limpios)

# Preprocesar todo el corpus
corpus_pre = [preprocesar_texto(doc) for doc in corpus_es]

print("\nEjemplo de preprocesamiento (primer documento):")
print("Original   :", corpus_es[0])
print("Preprocesamiento:", corpus_pre[0])


# CountVectorizer con texto limpio
# Aquí el vectorizador ya NO necesita convertir a minúsculas ni quitar acentos,
# porque eso lo hicimos manualmente.

vectorizer_count_pre = CountVectorizer(
    lowercase=False,
    strip_accents=None
)

X_count_pre = vectorizer_count_pre.fit_transform(corpus_pre)

imprimir_info_matriz("CountVectorizer - texto preprocesado", vectorizer_count_pre, X_count_pre)

# TfidfVectorizer con texto limpio

vectorizer_tfidf_pre = TfidfVectorizer(
    lowercase=False,
    strip_accents=None
)

X_tfidf_pre = vectorizer_tfidf_pre.fit_transform(corpus_pre)

imprimir_info_matriz("TfidfVectorizer - texto preprocesado", vectorizer_tfidf_pre, X_tfidf_pre)

# Comparación numérica (doc 0) preprocesado

comparar_fila(
    i=0,
    vectorizer_count=vectorizer_count_pre,
    X_count=X_count_pre,
    vectorizer_tfidf=vectorizer_tfidf_pre,
    X_tfidf=X_tfidf_pre,
    n_mostrar=15
)
```

"Al preprocesar el texto (minúsculas, sin acentos, sin stopwords) se reduce el tamaño del vocabulario."

"Las matrices de CountVectorizer y TfidfVectorizer con texto limpio suelen tener menos columnas y menos ruido."

"En CountVectorizer preprocesado, los conteos se concentran en palabras más informativas (se eliminaron artículos, preposiciones, etc.)

"En TfidfVectorizer preprocesado, los pesos TF-IDF destacan aún más términos relevantes porque se han eliminado muchos términos muy frecuentes pero poco útiles (stopwords)."

En el ejercicio 2 se repite el mismo procedimiento, pero ahora sobre el texto preprocesado: se pasan las reseñas a minúsculas, se eliminan acentos y se quitan stopwords en español antes de vectorizar. Como resultado, las matrices de **CountVectorize**r y **TfidfVectorizer** siguen teniendo las mismas dimensiones entre sí, pero el tamaño del vocabulario disminuye de 127 a 101 términos, por lo que las nuevas matrices pasan a tener forma (30, 101).

Al comparar las matrices del ejercicio 1 (texto crudo) con las del ejercicio 2 (texto preprocesado), **se observan dos diferencias generales**:

1.Menos columnas y menos ruido: al eliminar artículos, preposiciones y otras stopwords, desaparecen muchos tokens poco informativos; el vocabulario se hace más pequeño y las matrices son más compactas.

2.Información más relevante en los valores: en CountVectorizer preprocesado los conteos se concentran en palabras clave del contenido (por ejemplo “pelicula”, “actores”, “historia”, “emocionante”) y ya no en palabras funcionales; en TfidfVectorizer preprocesado los pesos TF‑IDF destacan aún más esos términos relevantes, porque se han eliminado muchos términos muy frecuentes pero semánticamente débiles.

**En resumen**, entre Count y TF‑IDF la principal diferencia es el tipo de valor que almacena cada celda (conteos vs pesos TF‑IDF), mientras que entre ejercicio 1 y ejercicio 2 la diferencia es el grado de limpieza del texto: el preprocesamiento reduce el vocabulario y hace que ambas representaciones se centren en palabras más representativas del sentimiento y contenido de las reseñas.
